```python
import openpyxl
wb=openpyxl.load_workbook("data/03_wage0.xlsx",data_only=True)#data_only只引用单元格数值
ws=wb.active
    
```

```python
rngs=list(ws.rows)
title=[c.value for c in rngs[0]]
#title的获得法二
#for r in range(ws.max_row,2,-1):
    #title=ws.cell(r,c,ws.cell(1,c).value for c in (1,8))
#print(title)
for r in range(ws.max_row,2,-1):
    ws.insert_rows(r)#idx=r#从下往上插入【插入默认在选定行之上】
    for c in range(1,8):#[1,8)
        ws.cell(r,c,ws.cell(1,c).value)#设置单元格内容[行号-注意是r，列号，内容]
wb.save("data/03_wagetest1.xlsx")
```
