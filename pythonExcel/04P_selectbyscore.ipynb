```python
import openpyxl
wb=openpyxl.load_workbook("data/02_grade0.xlsx")
ws=wb['各科成绩']
ws1=wb.create_sheet(title='总分大于300')

rngs=list(ws.rows)
ws1.append([c.value for c in list(ws.rows)[0]]+['总分'])

#获取数据区域
for row in rngs[1:]:#第一行不取
    #print(row)#单元格对象的形式
    #print([c.value for c in row])#各个单元格的数据
    l=[c.value for c in row]
    if sum([c.value for c in row][1:])>300:
        l=l+[sum([c.value for c in row][1:])]
        ws1.append(l)
           
wb.save('data/02_test1.xlsx')

```
